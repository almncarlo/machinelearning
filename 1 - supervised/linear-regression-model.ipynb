# Building the Linear Regression Model

#### What is regression?

Building a "best fit line" (supervised) for a set of data points. Using this linear model, we can predict values given an input that is mapped to an approximated output.



Let's start with a simple 2-D dataset to build a linear model.

#### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

#### Linear Model

Here's a random dataset of 3 points. Let's start by visualizing the data below using matplotlib.

```python
x_train = np.array([1.0, 1.5, 2.0])
y_train = np.array([200.0, 250.0, 300.0])

# plot data
plt.scatter(x_train, y_train, marker='^', c='r')

# plot title
plt.title("Sample Plot")

# x and y axes title
plt.ylabel('y')
plt.xlabel('x')

plt.show()
```

From the plot, the data points seem to follow a linear trend. If we wanted more granularity with our data but only have these three points, we can try modelling a function using linear regression.

Remember that the formula for a line is $y = wx + b$ where $w$ is the slope and $b$ is the y-intercept.

In linear regression, we use the following notation: $f(x) = wx + b$ where $w$ is the weight and $b$ is the bias.

Translating this into a function:

```python
def model(x, w, b):
    m = x.shape[0]
    f = np.zeros(m)
    for i in range(m):
        f[i] = w*x[i] + b
    return f
```

Let's use this function and the values $w=50$ and $b=0$ to plot a line on the graph.

```python
# initialize parameters
w = 50
b = 0

# initialize the linear model
f = model(x_train, w, b)

# plot the model
plt.plot(x_train, f, c='b', label='prediction model')

# plot actual values of the dataset
plt.scatter(x_train, y_train, marker='^', c='r', label='actual values')

# plot title
plt.title("Sample Plot")

# x and y axes title
plt.ylabel('y')
plt.xlabel('x')

plt.show()
```

From the graph, we can see that the line does not fit the data points. We can adjust the weight and bias for the blue line to better fit our data. Let's try to shift the line up by increasing the bias to $b=100$, and make the line steeper by changing the weight to be $w=100$.

```python
# initialize parameters
w = 100
b = 100

# initialize the linear model
f = model(x_train, w, b)

# plot the model
plt.plot(x_train, f, c='b', label='prediction model')

# plot actual values of the dataset
plt.scatter(x_train, y_train, marker='^', c='r', label='actual values')

# plot title
plt.title("Sample Plot")

# x and y axes title
plt.ylabel('y')
plt.xlabel('x')

plt.legend()
plt.show()
```

Now our linear prediction model fits our dataset better, and we can use this to approximate where each input x would map to an output y on the graph.
